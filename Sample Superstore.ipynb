# Data Visualization

```python
ship_mode_counts=sp1['Ship Mode'].value_counts().reset_index().rename(columns={'index':'Ship Mode','Ship Mode':'ship_mode_counts'})
ship_mode_counts
```

```python
plt.figure(figsize=(14,7))
sns.barplot(x=ship_mode_counts['Ship Mode'],y=ship_mode_counts['ship_mode_counts'],order=ship_mode_counts['Ship Mode'])
plt.xlabel("Ship Mode",fontsize=15)
plt.ylabel("Ship Mode Count",fontsize=15)
plt.title("Visualization of ship mode",fontsize=20)
plt.show()
```

```python
preferred_segment_type=sp1['Segment'].value_counts()
preferred_segment_type
```

```python
plt.figure(figsize=(14,7))
sns.countplot(x=sp1['Segment'],order=preferred_segment_type.index)
plt.xlabel("Type of segment",fontsize=15)
plt.ylabel("Count of segment",fontsize=15)
plt.title("Preferred Segment Type",fontsize=20)
plt.show()
```

```python
highest_sale=sp1.groupby('Category')['Sales'].mean().reset_index()
highest_sale
```

```python
plt.figure(figsize=(14,7))
sns.barplot(x=highest_sale['Category'],y=highest_sale['Sales'])
plt.xlabel("Category",fontsize=15)
plt.ylabel("Sales",fontsize=15)
plt.title("Categorywise Sales",fontsize=20)
plt.show()
```

```python
# create a loss dataframe
loss_sp=sp1[sp1['Profit']<0]
loss_sp
```

```python
loss_sp.shape
```

```python
loss_sp.info()
```

```python
# total loss
total_loss=np.negative(loss_sp['Profit'].sum())
print('Total_loss=%.2f' %total_loss)
```

```python
loss_in_segment=sp1.groupby('Segment').sum().reset_index()
loss_in_segment
```

```python
plt.figure(figsize=(14,7))
sns.barplot(x=loss_in_segment['Segment'],y=loss_in_segment['Profit'], data=loss_in_segment['Segment'])
plt.xlabel("Segment",fontsize=15)
plt.ylabel("Loss",fontsize=15)
plt.title("Segmentwise Loss")
plt.show()
```

```python
loss_in_state=loss_sp.groupby('State').sum().reset_index().sort_values('Profit', ascending=True).head(10)
loss_in_state
```

```python
plt.figure(figsize=(14,7))
sns.barplot(x=loss_in_state['State'],y=loss_in_state['Profit'],data=loss_in_state['State'])
plt.xlabel("Loss",fontsize=15)
plt.ylabel("State",fontsize=15)
plt.title("Statewise loss")
plt.show()
```
